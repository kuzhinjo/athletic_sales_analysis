```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv', low_memory=False)
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv', low_memory=False)
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020.head(-1)
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021.head(-1)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_2020_2021_sales_df = pd.concat([athletic_sales_2020, athletic_sales_2021], axis="rows", join="inner")
combined_2020_2021_sales_df.reset_index()
combined_2020_2021_sales_df.head(-1)
```

```python
# Check if any values are null.
combined_2020_2021_sales_df.isnull().sum()
```

```python
# Check the data type of each column
combined_2020_2021_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_2020_2021_sales_df["invoice_date"] = combined_2020_2021_sales_df.loc[:, "invoice_date"].astype('datetime64[ns]')
combined_2020_2021_sales_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_2020_2021_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
groupby_2020_2021_units_sold = combined_2020_2021_sales_df.groupby(['region', 'state', 'city']).agg({'units_sold':'sum'})
groupby_2020_2021_units_sold = groupby_2020_2021_units_sold.rename(columns={"units_sold": "Total_Products_Sold"}, level=0)

# Show the top 5 results.
sorted_groupby_2020_2021_units_sold = groupby_2020_2021_units_sold.sort_values(by='Total_Products_Sold', ascending=False)
sorted_groupby_2020_2021_units_sold.head(5)                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_2020_2021_units_sold = pd.pivot_table(combined_2020_2021_sales_df, 
                                 values='units_sold',
                                 index=['region', 'state', 'city'],
                                 aggfunc='sum')
# Rename the "units_sold" column to "Total_Products_Sold"
renamed_2020_2021_units_sold = pivot_2020_2021_units_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
sorted_groupby_2020_2021_units_sold = renamed_2020_2021_units_sold.sort_values(by='Total_Products_Sold', ascending=False)
sorted_groupby_2020_2021_units_sold.head(5)   
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
groupby_2020_2021_total_sales = combined_2020_2021_sales_df.groupby(['region', 'state', 'city']).agg({'total_sales':'sum'})
groupby_2020_2021_total_sales = groupby_2020_2021_total_sales.rename(columns={"total_sales": "Total Sales"}, level=0)


# Show the top 5 results.
sorted_groupby_2020_2021_total_sales = groupby_2020_2021_total_sales.sort_values(by='Total Sales', ascending=False)
sorted_groupby_2020_2021_total_sales.head(5) 
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_2020_2021_total_sales = pd.pivot_table(combined_2020_2021_sales_df, 
                                 values='total_sales',
                                 index=['region', 'state', 'city'],
                                 aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
renamed_2020_2021_total_sales = pivot_2020_2021_total_sales.rename(columns={"total_sales": "Total Sales"})


# Show the top 5 results.
sorted_groupby_2020_2021_total_sales = renamed_2020_2021_total_sales.sort_values(by='Total Sales', ascending=False)
sorted_groupby_2020_2021_total_sales.head(5)   
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
# joined_2020_2021_sales['product'].value_counts().head(10)
womens_ath_fw_df = combined_2020_2021_sales_df.copy()
womens_ath_fw_df["product"] = womens_ath_fw_df.loc[:, "product"].astype(str)

womens_ath_fw_df = womens_ath_fw_df.loc[womens_ath_fw_df['product'].str.contains('Women\'s Athletic Footwear', case = False)]
womens_ath_fw_df['retailer'].value_counts().head(10)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_ath_fw_units_sold_df = womens_ath_fw_df.groupby(['retailer', 'region', 'state', 'city']).agg({'units_sold':'sum'})
womens_ath_fw_units_sold_df = womens_ath_fw_units_sold_df.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, level=0)

# Show the top 5 results.
sorted_womens_ath_fw_units_sold_df = womens_ath_fw_units_sold_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
sorted_womens_ath_fw_units_sold_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_womens_ath_fw_units_sold = pd.pivot_table(combined_2020_2021_sales_df, 
                                 values='units_sold',
                                 index=['retailer', 'region', 'state', 'city'],
                                 aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
renamed_pivot_womens_ath_fw_units_sold = pivot_womens_ath_fw_units_sold.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
sorted_pivot_womens_ath_fw_units_sold = renamed_pivot_womens_ath_fw_units_sold.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
sorted_pivot_womens_ath_fw_units_sold.head(5) 
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_womens_ath_fw_total_sales_by_day = pd.pivot_table(womens_ath_fw_df, 
                                 values='total_sales',
                                 index=('invoice_date'),
                                 aggfunc='sum')
# Optional: Rename the "total_sales" column to "Total Sales"
renamed_pivot_womens_ath_fw_units_sold = pivot_womens_ath_fw_total_sales_by_day.rename(columns={"units_sold": "Total Sales"})

# Show the table.
print(renamed_pivot_womens_ath_fw_units_sold)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
total_sales_per_day = renamed_pivot_womens_ath_fw_units_sold.resample('D').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
total_sales_per_day.sort_values(by="total_sales", ascending=False)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
total_sales_per_week = renamed_pivot_womens_ath_fw_units_sold.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
total_sales_per_week.sort_values(by="total_sales", ascending=False)
```
